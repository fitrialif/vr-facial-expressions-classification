```python
%reset -f
%config IPCompleter.use_jedi = False

import os
import gc

import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
# 350 x 350
# (0=Angry, 1=Disgust, 2=Fear, 3=Happy, 4=Sad, 5=Surprise, 6=Neutral)
```

```python
csv_filepath = 'D:\\My Work\\VR\\dataset\\facial_expressions-master\\data\\legend.csv'

dataset_df = pd.read_csv(csv_filepath)

dataset_df
```

```python
emotions_map = {
    'anger':        0, 
    'surprise':     5, 
    'disgust':      1, 
    'fear':         2, 
    'neutral':      6, 
    'happiness':    3, 
    'sadness':      4,
    'contempt':     1, 
    'NEUTRAL':      6, 
    'SADNESS':      4, 
    'DISGUST':      1, 
    'FEAR':         2, 
    'SURPRISE':     5, 
    'ANGER':        0,
    'HAPPINESS':    3,
}

dataset_df.emotion = dataset_df.emotion.map(emotions_map)

dataset_df.drop(columns=['user.id'], inplace=True)

dataset_df = pd.concat([dataset_df, pd.Series(dtype=np.str, name='pixels')], 1)

dataset_df
```

```python
images_dshape = (350, 350)
images_dirpath = 'D:\\My Work\\VR\\dataset\\facial_expressions-master\\images'
```

```python
def grayimage2string(image: np.ndarray):
    return ' '.join(np.array(image.flatten(), np.str))


def read_image(image_path: str, dshape=None):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        return
    h, w = image.shape
    if dshape is not None and not (h == dshape[0] and w == dshape[1]):
        return
    return image


def read_image_as_string(image_path: str):
    image = read_image(image_path, dshape=images_dshape)
    if image is None:
        del image
        return np.NaN
    image_seq = grayimage2string(image)
    del image
    return image_seq


range_from = 13000
range_to = -1

dataset_df.pixels[range_from:range_to] = dataset_df.image[range_from:range_to].apply(
    lambda fname: read_image_as_string(os.path.join(images_dirpath, fname)))

dataset_df[range_from:range_to]
```
